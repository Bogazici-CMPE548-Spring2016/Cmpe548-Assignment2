### 1. (Importance Sampling)

```python
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt

p=0.5
x = np.linspace(-1,1,10000)
y_positive = np.power(1 -np.power(np.absolute(x),p),np.divide(1,p))
y_negative = -1*np.power(1 -np.power(np.absolute(x),p),np.divide(1,p))


plt.axis('equal')
plt.title('$|x|^p + |y|^p \leq 1$',fontsize=18)
plt.plot(x,y_positive,'k.',x,y_negative, 'k.')
plt.fill_between(x, y_positive, y_negative, where = y_positive >= y_negative, facecolor='black', interpolate=True)
plt.grid()
plt.show()
```

Assume $f(x)=y=[(1-|x|^p)^{1/p} \leq 1]$ and $p(x)=y= (1-|x|^p)^{1/p}$ and $q(x)=y=(1-|x|^2)^{0.5}$ which is suitable to get uniform sample, we can compute the area of the ball for $p=0.5$ via 

$2\dfrac{\sum\limits_{i=1}^N f(x^{(i)})w(x^{(i)})}{\sum\limits_{i=1}^N w(x^{(i)})}$ 

when

$w(x^{(i)}) = \dfrac{(1-|x^{(i)}|^p)^{1/p}}{(1-|x^{(i)}|^2)^{0.5}}$

```python
def sample_from_two_norm_ball(N):

    r = [np.sqrt(random.random()) for i in range(N)]
    theta = [random.random()*2*np.pi for i in range(N)]
    
    x = np.multiply(r,np.cos(theta)) # x = rcos(theta)
    y = np.multiply(r,np.sin(theta)) # y = rsin(theta)
    return x,y

def check_inequality(x,y):
    return np.power(np.absolute(y),p)+np.power(np.absolute(x),p) <= 1
    
```

```python
N = 1000
x,y = sample_from_two_norm_ball(N)
is_lower_than_one = check_inequality(x,y)
weighted_sum = 0.0
sum_of_weights = 0.0
for i in range(len(x)):
    weight = np.divide(np.power(1-np.power(np.absolute(x[i]),p),np.divide(1,p)),np.power(1-np.power(x[i],2),0.5))
    sum_of_weights += weight
    if is_lower_than_one[i]:
        weighted_sum += weight

area = 2*np.divide(weighted_sum,sum_of_weights)
print 'Area of the norm ball for p = 0.5 is ' + str(area) +' '+'when N = 1000'
        
        
    
```

Since variance of $f(x)w(x)$ is $<f^2(x)w^2(x)>_{q(x)} - {(<f(x)w(x)>_{q(x)})}^2$, the estimate variance can be calculated:

```python
expectaion_of_squares = 0.0
for i in range(len(x)):
    weight = np.divide(np.power(1-np.power(np.absolute(x[i]),p),np.divide(1,p)),np.power(1-np.power(x[i],2),0.5))
    if is_lower_than_one[i]:
        expectaion_of_squares += np.power(weight,2)    

variance = np.divide(expectaion_of_squares - np.power(area,2),N)
print 'Variance of the estimate is ' + str(variance) +' '+'when N = 1000'
```

```python
T = np.array([[0.25,0.25,0.25,0.25,0.25,0],[0,0.25,0,0,0.25,0],[0,0,0,0.5,0,0],[0.25,0.25,0.5,0,0.25,0.5],[0.25,0.25,0.25,0.25,0.25,0.25],[0.25,0,0,0,0,0.25]])
print 'Transition Matrix for the chain\n'
for row in T:
    print row
```

### (b)
The stationary distribution of $A$ is the normalized eigenvector of the transition matrix $T$, corresponding to the eigenvalue $\lambda = 1$

```python
eigen_values, eigen_vectors = np.linalg.eig(T)
i = np.argmax(np.isclose(eigen_values, 1.0))
v = eigen_vectors[:, i]
v = v / sum(v)
print v
```

### (c)
Detailed balance condition is satisfied when $T_{ji}\pi_i=T_{ij}\pi_j$, where $\pi$ is the stationary distribution and $T$ is the transition matrix.

```python
x = np.dot(T, np.diag(v))
np.isclose(x, x.T).all()
```

Thus, detailed balance condition is not satisfied for this process.

```python
pi = np.zeros(6)
pi[0] = 1
count = 0
while True:
    pi = np.dot(T,np.transpose(pi))
    diff = 0.0
    for j in range(len(pi)):
        diff += np.absolute(pi[j]-v[j])
        diff = np.divide(diff,0.5)
    if np.isclose(diff, 1e-08):
        print 'Mixing time is',count
        break
    else:
        count+=1

```

### (f)

```python
def simulate(T, N):
    chain = np.zeros(N).astype(int)
    chain[0] = 0
    for i in range(1,N):
        probability = T[:, chain[i-1]]
        chain[i] = np.random.choice(range(6), p = probability)
    return(chain)

nchains = 1000
chain_length = 50
samples = np.zeros([nchains, chain_length]).astype(int)
for i in range(0, nchains):
    ch = simulate(T, chain_length)
    samples[i,:] = ch
column_histograms = np.apply_along_axis(lambda x: np.histogram(x, bins=[0,1,2,3,4,5,6], density = True)[0], 0, samples)
plt.figure(figsize=(16,4))
sns.heatmap(column_histograms)
plt.show()
```
